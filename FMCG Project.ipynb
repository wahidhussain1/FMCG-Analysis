```python
import numpy as np
import pandas as pd
import datetime
import re as re
```

```python
df = pd.read_csv("C:\\Notes\\Python\\Project\\FMCG_Assignment_IOTA\\Raw Sales Data - Copy.csv")

df
```

```python
df.info()
```

```python
# Here we saw that, There are 224 rows and 7 columns and there is not a single null value in it and the
# data type of each row and column is object.
```

```python
df["Qty"] = df["Qty"].str.replace(",","").astype("int64")

df["Sales"] = df["Sales"].str.replace(",","").astype("int64")

df['Month'] = pd.to_datetime(df['Month'])

# Here we see that there are commas in the "Qty" and the "Sales" which are given,then we replaced them.
# After replacing we saw that the data types of 'Month','Qty' and'Sales' are given in the object,
# then we changed into 'datetime' and 'int'.
```

```python
df.info()

# We can see that the data types of 'Month','Qty' and'Sales' have been changed.
```

```python
df['Month-Year'] = df['Month'].dt.strftime('%b/%Y')
df
 
# Now We Created the new column of Month-year where we extracted the 'Month Name' and 'Year'.
```

```python
#1. Monthwise number of quantities sold..

MonthWise_Qty = pd.DataFrame(df.groupby('Month-Year')["Qty"].sum().reset_index())

MonthWise_Qty


# Here we extracted the 'Qty' based on 'Month' using 'dataframe' and 'groupby'.
```

```python
#2. Category wise Qty sold..


CategoryWise_Qty = pd.DataFrame(df.groupby('Category')['Qty'].sum().reset_index())

CategoryWise_Qty

# Now we extracted the 'Qty' based on 'Category'.
```

```python
#3. Sub-channel wise total sales

Sub_channelWise_sales = pd.DataFrame(df.groupby('Sub-Channels')["Sales"].sum().reset_index())

Sub_channelWise_sales


# Now we extracted the 'Sales' based on 'Sub-Channels'.
```

```python
# 4. Highest sale in which month


Highest_sales = pd.DataFrame(df.groupby('Month-Year')["Sales"].sum().reset_index())
Highest_sales = Highest_sales.sort_values('Sales',ascending=False).head(1)

Highest_sales


# Firstly we have extracted 'Sales' on the basis of 'Month' after extracting we sort the 'Sales' in descending order
# and we got the high sales of the month.
```

```python
df['Clean Product Name'] = df['Product Name'].str.replace('\d+','---').str.split('---').str[0]


# Here we removed the numbers from product name and created a new column with only text.
```

```python
# 5. Highest selling product (based on sales) (packaging quantity doesn't matter).

Highest_product = pd.DataFrame(df.groupby('Clean Product Name')["Sales"].sum().reset_index())


Highest_product = Highest_product.sort_values('Sales',ascending = False).head(1)

Highest_product

# Here we extracted sales based on 'product name' after extracting we sort the 'Sales' in descending order
# and we got the highest product based on sales.
```

```python
# 6. List of different products company is offering (packaging quantity doesn't matter). 

unique_product = pd.DataFrame(df['Clean Product Name'].value_counts().reset_index())

unique_product


# Here is list of the different company who is offering(packaging quantity doesn't matter).
```

```python
# 7. Create a single excel file to save all the outputs generated above in different subsheets.

writer = pd.ExcelWriter("C:\\Users\\hp\\Desktop\\Power BI\\FMCGi Project.xlsx")

MonthWise_Qty.to_excel(writer,sheet_name = "MonthWise Quantity",index=False)
CategoryWise_Qty.to_excel(writer,sheet_name = "CategoryWise Quantity",index=False)
Sub_channelWise_sales.to_excel(writer,sheet_name = "Sub-Channel Wise Sales",index=False)
Highest_sales.to_excel(writer,sheet_name = "Highest Sale",index=False)
Highest_product.to_excel(writer,sheet_name = "Highest Product",index=False)
unique_product.to_excel(writer,sheet_name = "Different Company",index=False)

writer.save()
```
